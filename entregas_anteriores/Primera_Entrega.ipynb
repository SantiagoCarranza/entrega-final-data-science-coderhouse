```python
# Valor del dolar de hoy
USD = 0.010
```

```python
def convertir_dolares(row):
    if row['Moneda'] == 'USD':
        return row['Precio']*USD
    elif row['Moneda'] == 'ARS':
        return row['Precio']*USD

df['Precio USD'] = df.apply(convertir_dolares, axis = 1)
```

Como valores límite para el mínimo y el máximo de los vehículos en el análisis se estableció $1000 y $120.000 dólares, ya que estos valores reflejan el rango en el que los valores reflejan un precio real y además dentro del rango en que quiere trabajar la empresa.
