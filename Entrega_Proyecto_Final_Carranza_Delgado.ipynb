## Revisión de duplicados

En este caso, todas las columnas podrían estar duplicadas excepto la columna ID, ya que se supone que este es un valor único para cada pulbicación en MercadoLibre.

```python
# Checkeo de duplicados para ID's
ids = df['ID']
n_duplicadas = len(df[ids.isin(ids[ids.duplicated()])].sort_values("ID"))
df[ids.isin(ids[ids.duplicated()])].sort_values("ID")
```

Podemos ver que hay 17.656 filas con ID's repetidos, incluso más de 2 veces, así que se eliminarán todas las filas con ID's repetidos.

```python
print(f"Largo actual: {len(df)}")
print(f"Filas duplicadas: {n_duplicadas}")
print(f"Filas esperadas después de la eliminación de duplicados: {len(df) - n_duplicadas}")
```

```python
df.drop_duplicates(subset = 'ID', inplace=True)
print(len(df))
```

```python
df.info()
```

```python
# A continuación se revisa el número de valores nulos por columna.
df.isnull().sum(axis = 0).sort_values(ascending=False)
```

De acuerdo a la información anterior podemos ver que hay varias columnas con muchos valores nan, por lo que se estableció como cota 50.000 valores nulos, las columnas con más de 50.000 nan serán eliminadas.

```python
serie_nulos = df.isnull().sum(axis = 0).sort_values(ascending=False)
serie_nulos
```

```python
# Se crea la variable columnas_filtradas que contiene las columnas con menos de 50.000 nan
columnas_filtradas = [x for x in serie_nulos[serie_nulos < 50000].keys().values]
columnas_filtradas
```

```python
# Se filtra el dataset
df = df[columnas_filtradas]
```

```python
# Valor del dolar de hoy
USD = 0.010
```

```python
def convertir_dolares(row):
    if row['Moneda'] == 'USD':
        return row['Precio']*USD
    elif row['Moneda'] == 'ARS':
        return row['Precio']*USD

df['Precio USD'] = df.apply(convertir_dolares, axis = 1)
```

Como valores límite para el mínimo y el máximo de los vehículos en el análisis se estableció $1.000  y $120.000 dólares, ya que estos valores reflejan el rango en el que los valores reflejan un precio real y además dentro del rango en que quiere trabajar la empresa.

```python
df = df[(df['Precio USD']<120000) & (df['Precio USD']>500)]
```

```python
len(df)
```

Después de estas modificaciones, nos quedamos con 86.981 datos, los que siguen siendo suficientes.
A continuación podemos ver cómo se distribuyen esos valores, respecto al precio.

```python
df['Precio USD'].hist(bins = 50, by=df['Condición'])
```

```python
def kilometrizador(row):
    if 'km' in row['Atributo Kilómetros']:
        return int(row['Atributo Kilómetros'].split(' ')[0])
    else:
        return 'NO EN KM'
```

```python
df['Kilómetros'] = df.apply(kilometrizador, axis = 1)
```

# Modelo 1: Árbol de decisión

```python
#Importamos el modelo
clf = DecisionTreeClassifier(max_depth = 13)

#Entrenamos el modelo
clf.fit(x_train,y_train)

#Prediccion en Train
y_pred_train = clf.predict(x_train)
#Prediccion en Test
y_pred_test = clf.predict(x_test)

#Calculo el accuracy en Train
acc_train = accuracy_score(y_true = y_train, y_pred = y_pred_train)


#Calculo el accuracy en Test
acc_test = accuracy_score(y_true = y_test, y_pred = y_pred_test)


print("Accuracy sobre TEST:", acc_test)

print("Accuracy sobre TRAIN:",acc_train)


resultados['Decision Tree Classifier'] = {"Test":acc_test, "Train":acc_train}
```

## Explicación parámetro

Los árboles de decisión no están regularizados por defecto. Esto significa que para obtener el menor error posible, pueden crear tantos nodos como necesiten. 

La regularización consiste en limitar de alguna forma las capacidades del modelo para obtener un modelo de aprendizaje automático que sea más simple y  generalice mejor.

Podemos usar varios hiper-parámetros para configurar cómo regularizamos los árboles de decisión.  En este caso veremos los más usados:

**max_depth:** la profundidad máxima del árbol. En el ejemplo anterior hemos usado max_depth = 13

*max.depth: profundidad máxima que pueden alcanzar los árboles. Por defecto (0) crece al máximo los árboles.
Se alcanza el numero maximo de niveles (“depth”) que debe tener el arbol, segun acordado por el usuario.
El arbol “mas grande” que se crea generalmente “sobreajusta”, es decir sigue mucho la tendencia de los datos. Las ultimas divisiones del arbol son las que generalmente causan el sobreajuste. Por lo tanto se requiere recortar el arbol.

## Validación y Optimización de Decision Tree

En esta etapa se identificara, desarrollara y aplicara diversos Metodos asociados a la mejora de los  modelos de Machine Learning. 

Uno de ellos es, el** métodos de validación**, también conocido como resampling, que son estrategias que permiten estimar la capacidad predictiva de los modelos cuando se aplican a nuevas observaciones, haciendo uso únicamente de los datos de entrenamiento.

El otro , es el  **metodo de Optimización de hiperparámetros**  que se utiliza  para determinar cuáles son los mejores hiperparámetros para el algoritmo de aprendizaje.



```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

skf = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
train, test = list(skf.split(x, y))[0]
x_train = x.iloc[train]
x_test = x.iloc[test]
y_train = y[train]
y_test= y[test]

dtc = DecisionTreeClassifier()

max_depth = list(np.arange(1, 22., step=3))
ccp_alpha = list(np.arange(0.0, 1., step=0.2))
param_grid = {"max_depth":max_depth, "ccp_alpha":ccp_alpha}


cv_car = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)
tunner = GridSearchCV(estimator=dtc, param_grid=param_grid, cv=cv_car)
tunner.fit(x_train, y_train)
```

```python
tunner_car.cv_results_
```

```python
# Con 2 parametros
print('Train score StratifiedKFold: ' + str(tunner_car.score(x_train, y_train)))
print('Test score StratifiedKFold: ' + str(tunner_car.score(x_test, y_test)))
```
