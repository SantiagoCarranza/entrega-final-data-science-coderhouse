Este cuaderno toma los datos descargados desde la api de mercadolibre, elimina los duplicados y crea un dataframe que guarda como CSV.

```python
import os
import json
```

```python
unicos = {}
```

```python
for each_file in os.listdir('saved_files'):
    print('saved_files/'+each_file)
    with open('saved_files/' + each_file) as file:
        guardado = file.read()
        for x in json.loads(guardado):
            unicos[x['id']] = x
```

```python
cada_registro = unicos['MLA932818798']
```

```python
base_de_datos = []
```

```python
contador = 1
for cada_key in unicos:
    
    cada_registro = unicos[cada_key]
    este_registro = {}
    este_registro['ID'] = cada_registro['id']
    este_registro['Titulo'] = cada_registro['title']
    try:
        este_registro['ID Vendedor'] = cada_registro['seller']['id']
        este_registro['ID Vendedor'] = cada_registro['seller']['id']
        este_registro['Link Vendedor'] = cada_registro['seller']['permalink']
        este_registro['Registro Vendedor'] = cada_registro['seller']['registration_date']
        este_registro['Vendedor de Autos'] = cada_registro['seller']['car_dealer']
        este_registro['Agencia Inmobiliaria'] = cada_registro['seller']['real_estate_agency']
    except:
        este_registro['ID Vendedor'] = 'indefinido'
        este_registro['ID Vendedor'] = 'indefinido'
        este_registro['Link Vendedor'] = 'indefinido'
        este_registro['Registro Vendedor'] = 'indefinido'
        este_registro['Vendedor de Autos'] = 'indefinido'
        este_registro['Agencia Inmobiliaria'] = 'indefinido'
    
    
    try: 
        este_registro['Transacciones Totales Vendedor'] = cada_registro['seller']['seller_reputation']['transactions']['total']
        este_registro['Valoraciones Positivas Vendedor'] = cada_registro['seller']['seller_reputation']['transactions']['ratings']['positive']
        este_registro['Valoraciones Negativas Vendedor'] = cada_registro['seller']['seller_reputation']['transactions']['ratings']['negative']
        este_registro['Valoraciones Neutrales Vendedor'] = cada_registro['seller']['seller_reputation']['transactions']['ratings']['neutral']
    except:
        este_registro['Transacciones Totales Vendedor'] = 'indefinido'
        este_registro['Valoraciones Positivas Vendedor'] = 'indefinido'
        este_registro['Valoraciones Negativas Vendedor'] = 'indefinido'
        este_registro['Valoraciones Neutrales Vendedor'] = 'indefinido'
    
    este_registro['Precio'] = cada_registro['price']
    este_registro['Moneda'] = cada_registro['currency_id']
    este_registro['Cantidad Disponible'] = cada_registro['available_quantity']
    este_registro['Unidades Vendidas'] = cada_registro['sold_quantity']
    este_registro['Tipo de Publicacin'] = cada_registro['listing_type_id']
    este_registro['Fecha Finalizacin'] = cada_registro['stop_time']
    este_registro['Condicin'] = cada_registro['condition']
    este_registro['Link'] = cada_registro['permalink']
    este_registro['Link Miniatura'] = cada_registro['thumbnail']
    este_registro['Acepta MercadoPago'] = cada_registro['accepts_mercadopago']
    este_registro['Estado'] = cada_registro['address']['state_name']
    este_registro['Ciudad'] = cada_registro['address']['city_name']
    este_registro['Latitud'] = cada_registro['location']['latitude']
    este_registro['Longitud'] = cada_registro['location']['longitude']
    try:
        este_registro['Pas Vendedor'] = cada_registro['seller_address']['country']['name']
        este_registro['Estado Vendedor'] = cada_registro['seller_address']['state']['name']
        este_registro['Ciudad Vendedor'] = cada_registro['seller_address']['city']['name']
    except:
        este_registro['Pas Vendedor'] = 'Indefinido'
        este_registro['Estado Vendedor'] = 'Indefinido'
        este_registro['Ciudad Vendedor'] = 'Indefinido'
    
    
    for cada_atributo in cada_registro['attributes']:
        este_registro['Atributo '+ cada_atributo['name']] = cada_atributo['value_name']
    base_de_datos.append(este_registro)
    if contador % 100 == 0:
        print(contador)
    contador+=1

```
